```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram
from qiskit.visualization import plot_state_city
%matplotlib inline
```

```python
###########################__________CIRCUIT__________###########################
```

```python
# CREATE A CIRCUIT

qr = QuantumRegister(2) # quantum registers
cr = ClassicalRegister(2) # classical registers

circuit = QuantumCircuit(qr, cr) # create circuit

circuit.draw(output='mpl')
```

```python
# QUANTUM GATES

circuit.h(qr[0]) # hadamard gate
circuit.x(qr[1]) # not gate
circuit.cx(qr[0], qr[1]) # cnot gate
circuit.cz(qr[0], qr[1]) # cz gate

circuit.barrier(qr[0]) # specific register barrier
circuit.barrier() # barrier

circuit.draw(output='mpl')
```

```python
# MEASUREMENTS

circuit.measure([qr[0], qr[1]], [cr[0], cr[1]]) # speceific measure
circuit.measure(qr, cr) # measure all

circuit.draw(output='mpl')
```

```python
###########################__________SIMULATION__________###########################
```

```python
# RUN CIRCUIT qasm_simulator

simulator_qasm_simulator = Aer.get_backend('qasm_simulator') # select machine
result_qasm_simulator = execute(circuit, backend = simulator_qasm_simulator, shots=1024).result() # execute circuit

plot_histogram(result_qasm_simulator.get_counts(circuit))
```

```python
# RUN CIRCUIT statevector_simulator

simulator_statevector_simulator = Aer.get_backend('statevector_simulator') # select machine
job = execute(circuit, backend = simulator_statevector_simulator, shots=1024).result() # execute circuit
result_statevector_simulator = job.get_statevector(circuit, decimals = 3)

print(result_statevector_simulator)
plot_state_city(result_statevector_simulator)
```
